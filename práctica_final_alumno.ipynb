```python
'''
8. Crea un nuevo dataset con los pokemon Legendary
'''
legendaries = df[df['Legendary']]
legendaries.reset_index(drop=True)
```

```python


field_names = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
lg_field_value = list(map(lambda field: legendaries[field].mean().round(2), field_names))


field_averages = lambda field: {(field[0] +' average'): field[1]}

legendary_list = list(map(field_averages, zip(field_names, lg_field_value)))
print(legendary_list)

lg_tuple = zip(field_names, lg_field_value)
lg_dct = dict((x,y) for x, y in lg_tuple)
lg_dct
```

```python
no_legendaries = df[df['Legendary']== False]

field_names = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
no_lg_field_value = list(map(lambda field: no_legendaries[field].mean().round(2), field_names))


field_averages = lambda field: {(field[0] + 'average'): field[1]}

no_legendary_list = list(map(field_averages, zip(field_names, no_lg_field_value)))
print(no_legendary_list)

no_lg_tuple = zip(field_names, no_lg_field_value)
no_lg_dct = dict((x,y) for x, y in no_lg_tuple)
no_lg_dct
```

```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


# A. Una función que calcule el precio total de la compra
total_compra = reduce(lambda total, item: total + articulos[item] if item in articulos else 0, compra, 0); print('Total compra:', total_compra)

# B. Una función que calcule el precio total de la compra con un descuento del 10%
desc_10 = lambda total_compra:  total_compra - (total_compra*0.01) ; print('Total con desc de 10%:' , desc_10(total_compra))

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
superior_100 = desc_10(total_compra) if total_compra > 100 else total_compra ; print('Compras superiores de 100: ', superior_100)

# D. Una función que calcule el precio total aplicando el IVA (21%)
total_iva = lambda total_compra:  total_compra + (total_compra*0.01) ; print('Total con IVA sin descuento:', total_iva(total_compra))

# E.lista los artículos cuyo precio es superior a 20€
superior_20 = list(filter(lambda item: True if item in articulos and articulos[item]>20 else False,compra)) ; print('Productos superiores a 20:', superior_20)
```
